```python
df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >= 10').sort_values('mean')
```

```python
ax= df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='bar',figsize=(12,5) ,title=('Top 10 Fastest Coasters'))
```

```python
ax= df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='bar',figsize=(12,5) ,title=('Top 10 Fastest Coasters'))
ax.set_ylabel('Avg Coaster Speed')
```
